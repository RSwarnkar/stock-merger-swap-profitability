```python
import scipy.stats as st
import numpy as np
# fitting part
samples = np.random.rand(10000)
dist = st.exponweib.fit(samples)
# generation part
sample_seed = np.random.rand(10000)
random_exponweib_samples = st.exponweib.ppf(sample_seed, *dist)
# plotting
import matplotlib.pyplot as plt
plt.hist(samples, label="uniform", alpha=.5)
plt.hist(random_exponweib_samples, label="exponweib", alpha=.5)
plt.legend()
plt.show()
```

```python
import pandas as pd 
import scipy.stats as st
import numpy as np

df  = pd.read_csv("dataset/HDFCBANK-ratio.csv")
_array = df["Ratio B"].to_numpy()
dist = st.exponweib.fit(_array)

print(str(dist))

n_days = 8000

# generation part
sample_seed = np.random.uniform(np.min(_array), np.max(_array), n_days)
random_exponweib_samples = st.exponweib.ppf(sample_seed, *dist)

# plotting
import matplotlib.pyplot as plt
plt.hist(random_exponweib_samples, label="exponweib")
plt.legend()
plt.show()
```
